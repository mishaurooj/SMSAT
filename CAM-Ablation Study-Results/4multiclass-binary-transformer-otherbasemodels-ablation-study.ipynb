# Base Models Ablation and k-Fold Support


1. Ablation Study for MFCC, Wavelet, Temporal, and All
2. Multi-class & Binary Metrics (Precision, Recall, F1, ROC-AUC)Confusion Matrix, Training & Testing Time Logging
3. Cross-ValidationSubject-Level K-Fold Cross-Validation
4. Excel Logging for all experiments


```python
import torch, torch.nn as nn, torch.optim as optim
import os, time, copy, librosa, numpy as np, pywt, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
from collections import Counter
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler, Subset
from sklearn.metrics import confusion_matrix, classification_report, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.model_selection import KFold, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from itertools import combinations

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
target_names = ["Music", "Normal(Silence)", "SpiritualMeditation"]
root_dir = r"C:\Users\muroo\Downloads\archive\ATS-data"

# =====================================
# 1. Dataset with Subject IDs & Feature Selection
# =====================================
class SMSATDataset(Dataset):
    def __init__(self, root_dir, feature_mode="all"):
        self.root_dir, self.feature_mode = root_dir, feature_mode
        self.file_paths, self.labels, self.subject_ids, self.class_counts = [], [], [], Counter()
        mapping = {"Music":0,"Normal(Silence)":1,"SpiritualMeditation":2}

        for cat in os.listdir(root_dir):
            path=os.path.join(root_dir,cat)
            if not os.path.isdir(path): continue
            label=mapping.get(cat,-1)
            for f in os.listdir(path):
                if f.endswith(".wav"):
                    self.file_paths.append(os.path.join(path,f))
                    self.labels.append(label)
                    self.subject_ids.append(f.split("_")[0])
                    self.class_counts[label]+=1

        total=sum(self.class_counts.values())
        self.class_weights={k:total/v for k,v in self.class_counts.items()}

    def __len__(self): return len(self.file_paths)

    def __getitem__(self, idx):
        wav,sr=librosa.load(self.file_paths[idx],sr=16000)
        mfcc=librosa.feature.mfcc(y=wav,sr=sr,n_mfcc=13).mean(axis=1)
        timef=np.array([librosa.feature.zero_crossing_rate(wav)[0].mean(),
                        librosa.feature.rms(y=wav)[0].mean()])
        coeffs=pywt.wavedec(wav,'db4',level=4)
        wavelet=np.array([s for c in coeffs for s in (np.mean(c),np.std(c))][:10])
        feats = mfcc if self.feature_mode=="mfcc" else \
                timef if self.feature_mode=="temporal" else \
                wavelet if self.feature_mode=="wavelet" else \
                np.concatenate([mfcc,timef,wavelet])
        return torch.tensor(feats,dtype=torch.float32), torch.tensor(self.labels[idx]), self.subject_ids[idx]

# =====================================
# 2. Models: BiLSTM & Transformer
# =====================================
class CalmnessBiLSTM(nn.Module):
    def __init__(self,input_dim=25,h=256,num_classes=3):
        super().__init__()
        self.lstm=nn.LSTM(input_dim,h,batch_first=True,bidirectional=True)
        self.fc1, self.fc2=nn.Linear(h*2,128), nn.Linear(128,num_classes)
        self.relu, self.drop=nn.ReLU(), nn.Dropout(0.3)
    def forward(self,x):
        x=x.unsqueeze(1); out,_=self.lstm(x)
        return self.fc2(self.drop(self.relu(self.fc1(out[:,-1,:]))))

class TimeSeriesTransformer(nn.Module):
    def __init__(self,input_dim=25,d_model=64,nhead=4,num_layers=2,num_classes=3):
        super().__init__()
        self.fc_in=nn.Linear(input_dim,d_model)
        enc=nn.TransformerEncoderLayer(d_model=d_model,nhead=nhead,dim_feedforward=128)
        self.enc=nn.TransformerEncoder(enc,num_layers=num_layers)
        self.fc_out=nn.Linear(d_model,num_classes)
    def forward(self,x):
        x=self.fc_in(x.unsqueeze(1))
        return self.fc_out(self.enc(x)[:,-1,:])

# =====================================
# 3. Training & Evaluation Utilities
# =====================================
def train_model(model,loader,epochs=350,lr=0.005):
    opt,loss_fn=optim.Adam(model.parameters(),lr=lr),nn.CrossEntropyLoss()
    best,bw=float("inf"),copy.deepcopy(model.state_dict())
    start=time.time()
    for ep in range(epochs):
        model.train(); total=correct=lsum=0
        for X,y,_ in loader:
            X,y=X.to(device),y.to(device)
            opt.zero_grad(); out=model(X)
            loss=loss_fn(out,y); loss.backward(); opt.step()
            lsum+=loss.item(); correct+=(out.argmax(1)==y).sum().item(); total+=len(y)
        print(f"[{ep+1}] Loss={lsum:.3f} Acc={100*correct/total:.2f}%")
        if lsum<best: best,bw=lsum,copy.deepcopy(model.state_dict())
    model.load_state_dict(bw)
    return model,time.time()-start

def eval_model(model,loader):
    model.eval(); ytrue,ypred,yprob=[],[],[]
    with torch.no_grad():
        for X,y,_ in loader:
            out=model(X.to(device)); prob=torch.softmax(out,1).cpu().numpy()
            ytrue+=y.numpy().tolist(); ypred+=out.argmax(1).cpu().numpy().tolist(); yprob+=prob.tolist()
    return np.array(ytrue),np.array(ypred),np.array(yprob)

def evaluate_binary(y,yp,pr):
    rows=[]
    for (i,j) in combinations(range(3),2):
        m=(y==i)|(y==j); yt=(y[m]==i).astype(int); ypbin=(yp[m]==i).astype(int); ys=pr[m][:,i]
        rows.append({"Pair":f"{target_names[i]} vs {target_names[j]}",
                     "Precision":precision_score(yt,ypbin),
                     "Recall":recall_score(yt,ypbin),
                     "F1":f1_score(yt,ypbin),
                     "ROC-AUC":roc_auc_score(yt,ys) if len(np.unique(yt))>1 else np.nan})
    df=pd.DataFrame(rows); print(df); return df

# =====================================
# 4. Subject-Level K-Fold CV (LOSO Removed)
# =====================================
def subject_kfold_crossval(dataset,model_class,input_dim=25,k=5,epochs=350):
    subs=np.unique(dataset.subject_ids); kf=KFold(n_splits=k,shuffle=True,random_state=42)
    rows=[]
    for tr,te in kf.split(subs):
        train_s=[subs[i] for i in tr]; test_s=[subs[i] for i in te]
        idx_tr=[i for i,v in enumerate(dataset.subject_ids) if v in train_s]
        idx_te=[i for i,v in enumerate(dataset.subject_ids) if v in test_s]
        dl_tr=DataLoader(Subset(dataset,idx_tr),batch_size=256,shuffle=True)
        dl_te=DataLoader(Subset(dataset,idx_te),batch_size=256,shuffle=False)
        m=model_class(input_dim=input_dim).to(device)
        m,tt=train_model(m,dl_tr,epochs=epochs)
        y,yp,_=eval_model(m,dl_te)
        rows.append({"FoldAcc":(y==yp).mean(),"F1":f1_score(y,yp,average="macro"),"TrainTime":tt})
    df=pd.DataFrame(rows); print("\nK-Fold CV Results:\n",df); return df

# =====================================
# 5. Classical ML Baselines
# =====================================
def extract_feats(ds):
    X,y=[],[]
    for i in range(len(ds)):
        f,l,_=ds[i]; X.append(f.numpy()); y.append(l.item())
    return np.array(X),np.array(y)

def classical_baselines():
    ds=SMSATDataset(root_dir,"all"); X,y=extract_feats(ds)
    Xtr,Xte,ytr,yte=train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)
    sc=StandardScaler(); Xtr=sc.fit_transform(Xtr); Xte=sc.transform(Xte)
    svm=SVC(kernel="rbf").fit(Xtr,ytr); rf=RandomForestClassifier(n_estimators=200).fit(Xtr,ytr)
    print("\nSVM:\n",classification_report(yte,svm.predict(Xte),target_names=target_names))
    print("\nRF:\n",classification_report(yte,rf.predict(Xte),target_names=target_names))
    return pd.DataFrame([{"Model":"SVM","Acc":(svm.predict(Xte)==yte).mean()},
                         {"Model":"RF","Acc":(rf.predict(Xte)==yte).mean()}])

# =====================================
# 6. Run Ablation + CV + Baselines + Transformer
# =====================================
abl_res=[]
for mode in ["all","mfcc","wavelet","temporal"]:
    print(f"\n=== Ablation: {mode} ===")
    ds=SMSATDataset(root_dir,mode)
    sw=[ds.class_weights[l] for l in ds.labels]
    dl=DataLoader(ds,batch_size=256,sampler=WeightedRandomSampler(sw,len(sw)))
    model=CalmnessBiLSTM(input_dim=(13 if mode=="mfcc" else 2 if mode=="temporal" else 10 if mode=="wavelet" else 25)).to(device)
    model,tt=train_model(model,dl,epochs=350)
    y,yp,pr=eval_model(model,dl)
    abl_res.append({"Feature":mode,"Acc":(y==yp).mean(),"TrainTime":tt})
    cm=confusion_matrix(y,yp); sns.heatmap(cm,annot=True,fmt="d",xticklabels=target_names,yticklabels=target_names); plt.title(f"CM-{mode}"); plt.show()
    print(classification_report(y,yp,target_names=target_names))
    evaluate_binary(y,yp,pr)

# Transformer Evaluation
print("\n=== Transformer on All Features ===")
dsT=SMSATDataset(root_dir,"all")
sw=[dsT.class_weights[l] for l in dsT.labels]
dlT=DataLoader(dsT,batch_size=256,sampler=WeightedRandomSampler(sw,len(sw)))
trans=TimeSeriesTransformer(25).to(device)
trans,tt=train_model(trans,dlT,epochs=350)
y,yp,pr=eval_model(trans,dlT)
print(classification_report(y,yp,target_names=target_names))
evaluate_binary(y,yp,pr)

# K-Fold 
kfold_df=subject_kfold_crossval(dsT,TimeSeriesTransformer,input_dim=25,k=3,epochs=350)

# Classical Baselines
baseline_df=classical_baselines()

# Save Excel
pd.ExcelWriter("final_unified_results.xlsx")
with pd.ExcelWriter("final_unified_results.xlsx") as w:
    pd.DataFrame(abl_res).to_excel(w,"Ablation",index=False)
    kfold_df.to_excel(w,"KFold",index=False)
    baseline_df.to_excel(w,"Baselines",index=False)
print("\n All results saved to final_unified_results.xlsx")
```

# Ablation for other ML models

```python
import os, time, librosa, numpy as np, pywt, pandas as pd
from collections import Counter
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from itertools import combinations

# ===============================
# Dataset
# ===============================
class SMSATDataset:
    def __init__(self, root_dir, feature_mode="all"):
        self.root_dir, self.feature_mode = root_dir, feature_mode
        self.file_paths, self.labels, self.subject_ids, self.class_counts = [], [], [], Counter()
        mapping = {"Music": 0, "Normal(Silence)": 1, "SpiritualMeditation": 2}

        for cat in os.listdir(root_dir):
            path = os.path.join(root_dir, cat)
            if not os.path.isdir(path): continue
            label = mapping.get(cat, -1)
            for f in os.listdir(path):
                if f.endswith(".wav"):
                    self.file_paths.append(os.path.join(path, f))
                    self.labels.append(label)
                    self.subject_ids.append(f.split("_")[0])
                    self.class_counts[label] += 1

    def __len__(self): return len(self.file_paths)

    def __getitem__(self, idx):
        wav, sr = librosa.load(self.file_paths[idx], sr=16000)
        mfcc = librosa.feature.mfcc(y=wav, sr=sr, n_mfcc=13).mean(axis=1)
        timef = np.array([librosa.feature.zero_crossing_rate(wav)[0].mean(),
                          librosa.feature.rms(y=wav)[0].mean()])
        coeffs = pywt.wavedec(wav, 'db4', level=4)
        wavelet = np.array([s for c in coeffs for s in (np.mean(c), np.std(c))][:10])

        if self.feature_mode == "mfcc":
            feats = mfcc
        elif self.feature_mode == "temporal":
            feats = timef
        elif self.feature_mode == "wavelet":
            feats = wavelet
        else:
            feats = np.concatenate([mfcc, timef, wavelet])
        return feats, self.labels[idx]

    def extract_features(self):
        X, y = [], []
        for i in range(len(self)):
            f, l = self[i]; X.append(f); y.append(l)
        return np.array(X), np.array(y)

# ===============================
# Pairwise Evaluation
# ===============================
def pairwise_evaluation(y_true, y_pred, y_prob, target_names):
    rows = []
    for (i, j) in combinations(range(3), 2):
        mask = (y_true == i) | (y_true == j)
        yt = (y_true[mask] == i).astype(int)
        yp = (y_pred[mask] == i).astype(int)
        ys = y_prob[mask][:, i]
        rows.append({
            "Pair": f"{target_names[i]} vs {target_names[j]}",
            "Precision": precision_score(yt, yp),
            "Recall": recall_score(yt, yp),
            "F1": f1_score(yt, yp),
            "ROC-AUC": roc_auc_score(yt, ys) if len(np.unique(yt)) > 1 else np.nan
        })
    return pd.DataFrame(rows)

# ===============================
# Train & Evaluate Function
# ===============================
def train_evaluate_model(model, Xtr, Xte, ytr, yte, model_name, feature_name, target_names):
    start = time.time()
    model.fit(Xtr, ytr)
    train_time = time.time() - start

    y_pred = model.predict(Xte)
    if hasattr(model, "predict_proba"):
        y_prob = model.predict_proba(Xte)
    else:  # For SVM without probability=True
        y_prob = np.zeros((len(yte), len(np.unique(yte))))

    acc = (y_pred == yte).mean()
    prec = precision_score(yte, y_pred, average="macro")
    rec = recall_score(yte, y_pred, average="macro")
    f1 = f1_score(yte, y_pred, average="macro")
    roc = roc_auc_score(pd.get_dummies(yte), pd.get_dummies(y_pred), multi_class="ovr")

    pairwise_df = pairwise_evaluation(yte, y_pred, y_prob, target_names)
    pairwise_df["Model"] = model_name
    pairwise_df["Feature"] = feature_name

    return {"Model": model_name, "Feature": feature_name, "Accuracy": acc, "Precision": prec,
            "Recall": rec, "F1": f1, "ROC-AUC": roc, "TrainTime": train_time}, pairwise_df

# ===============================
# Main Experiment
# ===============================
root_dir = r"C:\Users\muroo\Downloads\archive\ATS-data"
target_names = ["Music", "Normal(Silence)", "SpiritualMeditation"]
models = {
    "SVM": SVC(kernel="rbf", probability=True),
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "RF": RandomForestClassifier(n_estimators=200),
    "DT": DecisionTreeClassifier()
}

feature_modes = ["all", "mfcc", "wavelet", "temporal"]

results, pairwise_results = [], []

for fmode in feature_modes:
    dataset = SMSATDataset(root_dir, fmode)
    X, y = dataset.extract_features()
    Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
    sc = StandardScaler(); Xtr = sc.fit_transform(Xtr); Xte = sc.transform(Xte)

    for mname, m in models.items():
        print(f"\n=== {mname} on {fmode} features ===")
        res, p_df = train_evaluate_model(m, Xtr, Xte, ytr, yte, mname, fmode, target_names)
        results.append(res)
        pairwise_results.append(p_df)

# ===============================
# Save Results
# ===============================
results_df = pd.DataFrame(results)
pairwise_df = pd.concat(pairwise_results, ignore_index=True)

with pd.ExcelWriter("classical_ml_ablation_results.xlsx") as writer:
    results_df.to_excel(writer, sheet_name="MultiClass", index=False)
    pairwise_df.to_excel(writer, sheet_name="Pairwise", index=False)

print("\n Results saved to classical_ml_ablation_results.xlsx")
```

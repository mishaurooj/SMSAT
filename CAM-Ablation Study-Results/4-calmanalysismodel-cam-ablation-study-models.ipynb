# CalmAnalysisModel(CAM) Ablation Study

✔ Ablation Study (MFCC, Wavelet, Temporal, All)
✔ Multi-class & Binary Metrics (Precision, Recall, F1, ROC-AUC)
✔ Confusion Matrices & ROC Curves for each experiment
✔ LOSO & Subject-Level K-Fold CV
✔ Excel Logging with all results in one file
✔ LaTeX Tables for paper-ready results

```python
import torch
import torch.nn as nn
import torch.optim as optim
import os, time, copy, librosa, numpy as np, pywt, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler, Subset
from collections import Counter
from itertools import combinations
from sklearn.metrics import confusion_matrix, classification_report, precision_score, recall_score, f1_score, roc_auc_score, roc_curve, auc
from sklearn.model_selection import KFold
from sklearn.preprocessing import label_binarize

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
target_names = ["Music", "Normal(Silence)", "SpiritualMeditation"]
root_dir = r"C:\Users\muroo\Downloads\archive\ATS-data"

# =====================================================
# 1. Dataset (with Safe Loading)
# =====================================================
class SMSATDataset(Dataset):
    def __init__(self, root_dir, feature_mode="all"):
        self.root_dir, self.feature_mode = root_dir, feature_mode
        self.file_paths, self.labels, self.subject_ids, self.class_counts = [], [], [], Counter()
        mapping = {"Music": 0, "Normal(Silence)": 1, "SpiritualMeditation": 2}

        for cat in os.listdir(root_dir):
            path = os.path.join(root_dir, cat)
            if not os.path.isdir(path): 
                continue
            label = mapping.get(cat, -1)
            for f in os.listdir(path):
                if f.endswith(".wav"):
                    self.file_paths.append(os.path.join(path, f))
                    self.labels.append(label)
                    self.subject_ids.append(f.split("_")[0])
                    self.class_counts[label] += 1

        total = sum(self.class_counts.values())
        self.class_weights = {k: total/v for k, v in self.class_counts.items()}

    def __len__(self):
        return len(self.file_paths)

    def __getitem__(self, idx):
        try:
            wav, sr = librosa.load(self.file_paths[idx], sr=16000)
        except Exception as e:
            print(f"[ERROR] Failed to load {self.file_paths[idx]}: {e}")
            wav = np.zeros(16000)
            sr = 16000

        mfcc = librosa.feature.mfcc(y=wav, sr=sr, n_mfcc=13).mean(axis=1)
        timef = np.array([
            librosa.feature.zero_crossing_rate(wav)[0].mean(),
            librosa.feature.rms(y=wav)[0].mean()
        ])
        coeffs = pywt.wavedec(wav, 'db4', level=4)
        wavelet = np.array([s for c in coeffs for s in (np.mean(c), np.std(c))][:10])

        feats = (mfcc if self.feature_mode == "mfcc" else
                 timef if self.feature_mode == "temporal" else
                 wavelet if self.feature_mode == "wavelet" else
                 np.concatenate([mfcc, timef, wavelet]))

        return torch.tensor(feats, dtype=torch.float32), torch.tensor(self.labels[idx]), self.subject_ids[idx]

# =====================================================
# 2. Model (BiLSTM)
# =====================================================
class CalmnessAnalysisModel(nn.Module):
    def __init__(self, input_dim=25, hidden_dim=256, num_classes=3):
        super().__init__()
        self.lstm = nn.LSTM(input_dim, hidden_dim, batch_first=True, bidirectional=True)
        self.fc1, self.fc2 = nn.Linear(hidden_dim * 2, 128), nn.Linear(128, num_classes)
        self.relu, self.drop = nn.ReLU(), nn.Dropout(0.3)

    def forward(self, x):
        x = x.unsqueeze(1)
        out, _ = self.lstm(x)
        x = self.relu(self.fc1(out[:, -1, :]))
        return self.fc2(self.drop(x))

# =====================================================
# 3. Training & Evaluation Utilities
# =====================================================
def train_model(model, loader, epochs=350, lr=0.005):
    opt, loss_fn = optim.Adam(model.parameters(), lr=lr), nn.CrossEntropyLoss()
    best_loss, best_w = float("inf"), copy.deepcopy(model.state_dict())
    start = time.time()
    for ep in range(epochs):
        model.train()
        total = correct = lsum = 0
        for X, y, _ in loader:
            X, y = X.to(device), y.to(device)
            opt.zero_grad()
            out = model(X)
            loss = loss_fn(out, y)
            loss.backward()
            opt.step()
            lsum += loss.item()
            correct += (out.argmax(1) == y).sum().item()
            total += len(y)
        print(f"[{ep+1}] Loss={lsum:.3f} Acc={100*correct/total:.2f}%")
        if lsum < best_loss:
            best_loss, best_w = lsum, copy.deepcopy(model.state_dict())
    model.load_state_dict(best_w)
    return model, (time.time() - start)  # returns seconds

def eval_model(model, loader):
    model.eval()
    ytrue, ypred, yprob = [], [], []
    with torch.no_grad():
        for X, y, _ in loader:
            out = model(X.to(device))
            prob = torch.softmax(out, 1).cpu().numpy()
            ytrue += y.numpy().tolist()
            ypred += out.argmax(1).cpu().numpy().tolist()
            yprob += prob.tolist()
    return np.array(ytrue), np.array(ypred), np.array(yprob)

# =====================================================
# Binary & ROC Evaluation
# =====================================================
def evaluate_binary(y, yp, pr):
    rows = []
    for (i, j) in combinations(range(3), 2):
        m = (y == i) | (y == j)
        yt = (y[m] == i).astype(int)
        ypbin = (yp[m] == i).astype(int)
        ys = pr[m][:, i]
        rows.append({
            "Pair": f"{target_names[i]} vs {target_names[j]}",
            "Precision": precision_score(yt, ypbin),
            "Recall": recall_score(yt, ypbin),
            "F1": f1_score(yt, ypbin),
            "ROC-AUC": roc_auc_score(yt, ys) if len(np.unique(yt)) > 1 else np.nan
        })
    return pd.DataFrame(rows)

# =====================================================
# 4. LOSO & K-Fold CV (with Time in Minutes)
# =====================================================
def loso_crossval(dataset, model_class, input_dim=25, epochs=350):
    subs = np.unique(dataset.subject_ids)
    res = []
    for s in subs:
        idx_tr = [i for i, v in enumerate(dataset.subject_ids) if v != s]
        idx_te = [i for i, v in enumerate(dataset.subject_ids) if v == s]
        dl_tr = DataLoader(Subset(dataset, idx_tr), batch_size=256, shuffle=True, pin_memory=True, num_workers=0)
        dl_te = DataLoader(Subset(dataset, idx_te), batch_size=256, shuffle=False, pin_memory=True, num_workers=0)
        m = model_class(input_dim=input_dim).to(device)
        m, tt = train_model(m, dl_tr, epochs=epochs)
        y, yp, _ = eval_model(m, dl_te)
        res.append({"Subject": s, "Acc": (y == yp).mean(), "F1": f1_score(y, yp, average="macro"), "TrainTime_min": tt / 60})
    return pd.DataFrame(res)

def subject_kfold_crossval(dataset, model_class, input_dim=25, k=5, epochs=350):
    subs = np.unique(dataset.subject_ids)
    kf = KFold(n_splits=k, shuffle=True, random_state=42)
    rows = []
    for tr, te in kf.split(subs):
        tr_s = [subs[i] for i in tr]
        te_s = [subs[i] for i in te]
        idx_tr = [i for i, v in enumerate(dataset.subject_ids) if v in tr_s]
        idx_te = [i for i, v in enumerate(dataset.subject_ids) if v in te_s]
        dl_tr = DataLoader(Subset(dataset, idx_tr), batch_size=256, shuffle=True, pin_memory=True, num_workers=0)
        dl_te = DataLoader(Subset(dataset, idx_te), batch_size=256, shuffle=False, pin_memory=True, num_workers=0)
        m = model_class(input_dim=input_dim).to(device)
        m, tt = train_model(m, dl_tr, epochs=epochs)
        y, yp, _ = eval_model(m, dl_te)
        rows.append({"FoldAcc": (y == yp).mean(), "F1": f1_score(y, yp, average="macro"), "TrainTime_min": tt / 60})
    return pd.DataFrame(rows)

# =====================================================
# 5. Ablation Study
# =====================================================
if __name__ == "__main__":
    abl_res = []
    for mode in ["all", "mfcc", "wavelet", "temporal"]:
        print(f"\n=== Running Ablation: {mode} ===")
        ds = SMSATDataset(root_dir, mode)
        sw = [ds.class_weights[l] for l in ds.labels]
        dl = DataLoader(ds, batch_size=256, sampler=WeightedRandomSampler(sw, len(sw)), pin_memory=True, num_workers=0)
        model = CalmnessAnalysisModel(input_dim=(13 if mode == "mfcc" else 2 if mode == "temporal" else 10 if mode == "wavelet" else 25)).to(device)
        model, tt = train_model(model, dl, epochs=350)
        y, yp, pr = eval_model(model, dl)
        bin_df = evaluate_binary(y, yp, pr)
        abl_res.append({"Feature": mode, "Acc": (y == yp).mean(), "F1": f1_score(y, yp, average="macro"), "TrainTime_min": tt / 60})
        print(classification_report(y, yp, target_names=target_names))
        print(bin_df)
        cm = confusion_matrix(y, yp)
        sns.heatmap(cm, annot=True, fmt="d", xticklabels=target_names, yticklabels=target_names)
        plt.title(f"Confusion Matrix - {mode}")
        plt.show()

    # =====================================================
    # 6. LOSO & K-Fold CV
    # =====================================================
    ds_all = SMSATDataset(root_dir, "all")
    print("LOSO")
    loso_df = loso_crossval(ds_all, CalmnessAnalysisModel, input_dim=25, epochs=350)
    print("K-Fold CV")
    kfold_df = subject_kfold_crossval(ds_all, CalmnessAnalysisModel, input_dim=25, k=3, epochs=350)

    # =====================================================
    # 7. Save Results with Overall Mean
    # =====================================================
    output_path = os.path.join(root_dir, "final_results_with_all_metrics.xlsx")
    abl_df = pd.DataFrame(abl_res)

    abl_summary = pd.DataFrame([{"Feature": "Overall Mean", "Acc": abl_df["Acc"].mean(),
                                 "F1": abl_df["F1"].mean(), "TrainTime_min": abl_df["TrainTime_min"].mean()}])
    loso_summary = pd.DataFrame([{"Subject": "Overall Mean", "Acc": loso_df["Acc"].mean(),
                                  "F1": loso_df["F1"].mean(), "TrainTime_min": loso_df["TrainTime_min"].mean()}])
    kfold_summary = pd.DataFrame([{"FoldAcc": kfold_df["FoldAcc"].mean(), "F1": kfold_df["F1"].mean(),
                                   "TrainTime_min": kfold_df["TrainTime_min"].mean()}])

    with pd.ExcelWriter(output_path) as w:
        pd.concat([abl_df, abl_summary], ignore_index=True).to_excel(w, "Ablation", index=False)
        pd.concat([loso_df, loso_summary], ignore_index=True).to_excel(w, "LOSO", index=False)
        pd.concat([kfold_df, kfold_summary], ignore_index=True).to_excel(w, "KFold", index=False)

    print(f"\n✅ All experiments completed. Results saved to {output_path} (with summary rows)")
```

# Layer wise + Feature Ablation Study

```python
import torch
import torch.nn as nn
import torch.optim as optim
import os, time, copy, librosa, numpy as np, pywt, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from collections import Counter
from itertools import combinations
from sklearn.metrics import confusion_matrix, classification_report, f1_score, precision_score, recall_score, roc_auc_score, accuracy_score

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
target_names = ["Music", "Normal(Silence)", "SpiritualMeditation"]
root_dir = r"C:\Users\muroo\Downloads\archive\ATS-data"

# =====================================================
# 1. Dataset
# =====================================================
class SMSATDataset(Dataset):
    def __init__(self, root_dir, feature_mode="all"):
        self.root_dir, self.feature_mode = root_dir, feature_mode
        self.file_paths, self.labels, self.subject_ids, self.class_counts = [], [], [], Counter()
        mapping = {"Music": 0, "Normal(Silence)": 1, "SpiritualMeditation": 2}

        for cat in os.listdir(root_dir):
            path = os.path.join(root_dir, cat)
            if not os.path.isdir(path): continue
            label = mapping.get(cat, -1)
            for f in os.listdir(path):
                if f.endswith(".wav"):
                    self.file_paths.append(os.path.join(path, f))
                    self.labels.append(label)
                    self.subject_ids.append(f.split("_")[0])
                    self.class_counts[label] += 1

        total = sum(self.class_counts.values())
        self.class_weights = {k: total / v for k, v in self.class_counts.items()}

    def __len__(self): return len(self.file_paths)

    def __getitem__(self, idx):
        try:
            wav, sr = librosa.load(self.file_paths[idx], sr=16000)
        except Exception as e:
            print(f"[ERROR] Failed to load {self.file_paths[idx]}: {e}")
            wav = np.zeros(16000); sr = 16000

        mfcc = librosa.feature.mfcc(y=wav, sr=sr, n_mfcc=13).mean(axis=1)
        timef = np.array([librosa.feature.zero_crossing_rate(wav)[0].mean(),
                          librosa.feature.rms(y=wav)[0].mean()])
        coeffs = pywt.wavedec(wav, 'db4', level=4)
        wavelet = np.array([s for c in coeffs for s in (np.mean(c), np.std(c))][:10])

        feats = (mfcc if self.feature_mode == "mfcc" else
                 timef if self.feature_mode == "temporal" else
                 wavelet if self.feature_mode == "wavelet" else
                 np.concatenate([mfcc, timef, wavelet]))

        return torch.tensor(feats, dtype=torch.float32), torch.tensor(self.labels[idx]), self.subject_ids[idx]

# =====================================================
# 2. Model with Layer Ablation
# =====================================================
class CalmnessAnalysisModel(nn.Module):
    def __init__(self, input_dim=25, hidden_dim=256, num_classes=3, use_lstm=True, use_fc1=True, use_dropout=True):
        super().__init__()
        self.use_lstm = use_lstm
        self.use_fc1 = use_fc1
        self.use_dropout = use_dropout

        if use_lstm:
            self.lstm = nn.LSTM(input_dim, hidden_dim, batch_first=True, bidirectional=True)
            lstm_out_dim = hidden_dim * 2
        else:
            lstm_out_dim = input_dim

        if use_fc1:
            self.fc1 = nn.Linear(lstm_out_dim, 128)
            self.fc2 = nn.Linear(128, num_classes)
        else:
            self.fc2 = nn.Linear(lstm_out_dim, num_classes)

        self.relu = nn.ReLU()
        self.drop = nn.Dropout(0.3)

    def forward(self, x):
        x = x.unsqueeze(1)
        if self.use_lstm:
            out, _ = self.lstm(x)
            x = out[:, -1, :]
        else:
            x = x.squeeze(1)

        if self.use_fc1:
            x = self.relu(self.fc1(x))
            if self.use_dropout:
                x = self.drop(x)
        return self.fc2(x)

# =====================================================
# 3. Training & Evaluation
# =====================================================
def train_model(model, loader, epochs=350, lr=0.005):
    opt, loss_fn = optim.Adam(model.parameters(), lr=lr), nn.CrossEntropyLoss()
    best_loss, best_w = float("inf"), copy.deepcopy(model.state_dict())
    start = time.time()

    for ep in range(epochs):
        model.train(); total = correct = lsum = 0
        for X, y, _ in loader:
            X, y = X.to(device), y.to(device)
            opt.zero_grad()
            out = model(X)
            loss = loss_fn(out, y)
            loss.backward(); opt.step()
            lsum += loss.item(); correct += (out.argmax(1) == y).sum().item(); total += len(y)
        print(f"[{ep+1}] Loss={lsum:.3f} Acc={100*correct/total:.2f}%")
        if lsum < best_loss:
            best_loss, best_w = lsum, copy.deepcopy(model.state_dict())
    model.load_state_dict(best_w)
    return model, (time.time() - start)

def eval_model(model, loader):
    model.eval(); ytrue, ypred, yprob = [], [], []
    with torch.no_grad():
        for X, y, _ in loader:
            out = model(X.to(device))
            prob = torch.softmax(out, 1).cpu().numpy()
            ytrue += y.numpy().tolist()
            ypred += out.argmax(1).cpu().numpy().tolist()
            yprob += prob.tolist()
    return np.array(ytrue), np.array(ypred), np.array(yprob)

# =====================================================
# 4. Pairwise Binary Metrics
# =====================================================
def evaluate_pairwise_metrics(ytrue, ypred, yprob):
    results = []
    classes = np.unique(ytrue)
    for (i, j) in combinations(classes, 2):
        mask = (ytrue == i) | (ytrue == j)
        y_bin = (ytrue[mask] == i).astype(int)
        yp_bin = (ypred[mask] == i).astype(int)
        prob_bin = yprob[mask][:, i]

        results.append({
            "ClassPair": f"{target_names[i]} vs {target_names[j]}",
            "Accuracy": accuracy_score(y_bin, yp_bin),
            "Precision": precision_score(y_bin, yp_bin, zero_division=0),
            "Recall": recall_score(y_bin, yp_bin, zero_division=0),
            "F1": f1_score(y_bin, yp_bin, zero_division=0),
            "ROC-AUC": roc_auc_score(y_bin, prob_bin) if len(np.unique(y_bin)) > 1 else np.nan
        })
    return pd.DataFrame(results)

# =====================================================
# 5. Layer + Feature Ablation Study (No K-Fold)
# =====================================================
if __name__ == "__main__":
    layer_settings = [
        {"name": "FullModel", "use_lstm": True, "use_fc1": True, "use_dropout": True},
        {"name": "NoLSTM", "use_lstm": False, "use_fc1": True, "use_dropout": True},
        {"name": "NoFC1", "use_lstm": True, "use_fc1": False, "use_dropout": True},
        {"name": "NoDropout", "use_lstm": True, "use_fc1": True, "use_dropout": False}
    ]

    abl_res = []
    pairwise_all = []

    for mode in ["all", "mfcc", "wavelet", "temporal"]:
        ds = SMSATDataset(root_dir, mode)
        sw = [ds.class_weights[l] for l in ds.labels]
        dl = DataLoader(ds, batch_size=256, sampler=WeightedRandomSampler(sw, len(sw)))

        for setting in layer_settings:
            print(f"\n=== Running Ablation: Feature={mode}, Model={setting['name']} ===")
            model = CalmnessAnalysisModel(
                input_dim=(13 if mode == "mfcc" else 2 if mode == "temporal" else 10 if mode == "wavelet" else 25),
                use_lstm=setting["use_lstm"],
                use_fc1=setting["use_fc1"],
                use_dropout=setting["use_dropout"]
            ).to(device)

            model, tt = train_model(model, dl, epochs=350)
            y, yp, ypr = eval_model(model, dl)

            # 🔹 Multi-class metrics
            acc = (y == yp).mean()
            f1 = f1_score(y, yp, average="macro")
            print("\n--- Multi-Class Classification Report ---")
            print(classification_report(y, yp, target_names=target_names))

            # 🔹 Pairwise metrics
            pairwise_df = evaluate_pairwise_metrics(y, yp, ypr)
            print("\n--- Pairwise Binary Metrics ---")
            print(pairwise_df)

            # 🔹 Store results
            abl_res.append({"Feature": mode, "ModelVariant": setting["name"], "Acc": acc, "F1": f1, "TrainTime_min": tt / 60})
            pairwise_all.append(pairwise_df.assign(Feature=mode, ModelVariant=setting["name"]))

            # 🔹 Confusion matrix
            cm = confusion_matrix(y, yp)
            sns.heatmap(cm, annot=True, fmt="d", xticklabels=target_names, yticklabels=target_names)
            plt.title(f"Confusion Matrix - {mode} - {setting['name']}")
            plt.show()

    # =====================================================
    # 6. Save Results (Ablation + Pairwise)
    # =====================================================
    output_path = os.path.join(root_dir, "final_layer_feature_ablation_with_pairwise.xlsx")
    abl_df = pd.DataFrame(abl_res)
    pairwise_results = pd.concat(pairwise_all, ignore_index=True)

    summary = pd.DataFrame([{
        "Feature": "OverallMean", "ModelVariant": "All",
        "Acc": abl_df["Acc"].mean(), "F1": abl_df["F1"].mean(),
        "TrainTime_min": abl_df["TrainTime_min"].mean()
    }])

    with pd.ExcelWriter(output_path) as w:
        pd.concat([abl_df, summary], ignore_index=True).to_excel(w, "LayerFeatureAblation", index=False)
        pairwise_results.to_excel(w, "PairwiseMetrics", index=False)

    print(f"\n✅ Experiments completed. Results saved to {output_path}")
```
